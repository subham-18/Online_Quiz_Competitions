```python
import random

# List of questions and answers
questions = [
    {
        "question": "What is the capital of France?",
        "options": ["A. Paris", "B. Rome", "C. Madrid", "D. London"],
        "answer": "A"
    },
    {
        "question": "Who painted the Mona Lisa?",
        "options": ["A. Leonardo da Vinci", "B. Vincent van Gogh", "C. Pablo Picasso", "D. Claude Monet"],
        "answer": "A"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["A. Mars", "B. Venus", "C. Jupiter", "D. Saturn"],
        "answer": "A"
    },
     {
        "question": "What is the largest ocean in the world?",
        "options": ["A. Pacific Ocean", "B. Atlantic Ocean", "C. Indian Ocean", "D. Arctic Ocean"],
        "answer": "A"
    },
    {
        "question": "What is the chemical symbol for gold?",
        "options": ["A. Au", "B. Ag", "C. Fe", "D. Pb"],
        "answer": "A"
    },
    {
        "question": "Who wrote the novel 'Pride and Prejudice'?",
        "options": ["A. Jane Austen", "B. Charles Dickens", "C. William Shakespeare", "D. F. Scott Fitzgerald"],
        "answer": "A"
    },
    {
        "question": "What is the capital of Australia?",
        "options": ["A. Sydney", "B. Melbourne", "C. Canberra", "D. Perth"],
        "answer": "C"
    },
    {
        "question": "Who is the author of the Harry Potter book series?",
        "options": ["A. J.K. Rowling", "B. Stephen King", "C. Dan Brown", "D. George R.R. Martin"],
        "answer": "A"
    },
    {
        "question": "Which country won the FIFA World Cup 2018?",
        "options": ["A. France", "B. Brazil", "C. Germany", "D. Argentina"],
        "answer": "A"
    },
    {
        "question": "What is the chemical symbol for water?",
        "options": ["A. H2O", "B. CO2", "C. O2", "D. NaCl"],
        "answer": "A"
    },
    {
        "question": "Who painted the famous painting 'The Starry Night'?",
        "options": ["A. Vincent van Gogh", "B. Pablo Picasso", "C. Leonardo da Vinci", "D. Claude Monet"],
        "answer": "A"
    }
]

# Function to conduct the quiz competition
def conduct_quiz():
    players = {}  # Dictionary to store players and their scores
    
    num_players = int(input("Enter the number of players: "))
    
    # Registering players and initializing scores
    for i in range(num_players):
        player_name = input(f"Enter the name of Player {i+1}: ")
        players[player_name] = 0
    
    # Conducting the quiz
    for question in questions:
        random.shuffle(question["options"])  # Randomize the order of options
        print("\n")
        print(question["question"])
        for option in question["options"]:
            print(option)
        
        # Get player answers
        for player in players.keys():
            answer = input(f"\n{player}, enter your answer (A, B, C, D): ")
            if answer.upper() == question["answer"]:
                players[player] += 1
    
    # Displaying final scores
    print("\nQuiz competition results:")
    for player, score in players.items():
        print(f"{player}: {score} points")

# Run the quiz competition
conduct_quiz()
```
